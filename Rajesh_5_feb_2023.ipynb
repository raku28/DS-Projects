## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue. 

### Ans 5

```python
def calculate_revenue(discounted_price):
    if discounted_price > 600:
        return 0.25 * discounted_price
    elif discounted_price > 350 and discounted_price <= 600:
        return 0.15 * discounted_price
    elif discounted_price > 100 and discounted_price <= 350:
        return 0.10 * discounted_price
    else:
        return 0.05 * discounted_price


df['Revenue'] = df['discounted_price'].apply(calculate_revenue)


total_revenue = df['Revenue'].sum()
print(total_revenue)
df.head()
```

## Q 8. Create a new dataframe to include the Brand specific information as stated:
## i.	total number of orders placed per Brand
## ii.	total retail_price per Brand
## iii.	total discounted_price per Brand, and 
## iv.	total BrandRevenue generated per Brand. 

## Also, draw a pairplot using these four features. (6 marks)


### Ans 8

```python
grouped = df.groupby('Brand')
total_orders = grouped['Customer_uniq_id'].count()
print(total_orders)
total_retail_price = grouped['retail_price'].sum()
print(total_retail_price)
total_discounted_price = grouped['discounted_price'].sum()
print(total_discounted_price)
total_brand_revenue = total_discounted_price - grouped['Revenue'].sum()
print(total_brand_revenue)
brand_df = pd.DataFrame({'Total_Orders': total_orders,
                        'Total_Retail_Price': total_retail_price,
                        'Total_Discounted_Price': total_discounted_price,
                        'Total_Brand_Revenue': total_brand_revenue})
brand_df.reset_index(inplace=True)
sns.pairplot(brand_df)
```

## Q 9.Compare performance regionwise

## a)	Draw a lineplot for the monthly Revenue of E-Commerce Company for each region separately.  (4 marks) 

## b)	Identify the best and the worst performing months for each region. (2 marks)

#### Note: Only those days with actual orders(Order_Date) placed are present in the dataset. Assuming there were no orders on other days

### Ans 9 a)

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['month'] = df['Order_Date'].dt.month
grouped = df.groupby(['Region', 'month'])['Revenue'].sum().reset_index()
sns.lineplot(x='month', y='Revenue', hue='Region', data=grouped)
plt.title('Monthly Revenue of E-Commerce Company for each Region')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.show()
```

### Ans 9 b)

```python
best_months = grouped.groupby('Region')['Revenue'].idxmax()

worst_months = grouped.groupby('Region')['Revenue'].idxmin()


print('Best performing months for each region:')
print(best_months)

print('\nWorst performing months for each region:')
print(worst_months)
```
